## K means clustering

```python
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
object_name = df['Object Name']
df.drop('Object Name', axis=1, inplace=True)
```

```python
column_names = df.columns
```

```python
column_names
```

```python
for column in column_names:
    scaler = MinMaxScaler()
    scaler.fit(df[column].values.reshape(-1, 1))
    df[column] = scaler.transform(df[column].values.reshape(-1, 1))
```

```python
df
```

```python
# Feature Selection
features = ['Period Start','Possible Impacts', 'Asteroid Velocity', 'Asteroid Magnitude', 'Asteroid Diameter (km)' ,'Cumulative Palermo Scale']
```

```python
k_rng = range(2, 20)
ssd = [] #Sum of squared distance
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    ssd.append(km.inertia_)
```

```python
ssd
```

```python
plt.xlabel("K")
plt.ylabel("Sum of squared error")
plt.plot(k_rng, ssd, 'o--')
```

```python
pd.Series(ssd).diff()
```

From the Chart 6 and 11 look good K values

```python
model = KMeans(n_clusters=6)
y_pred = model.fit_predict(df)
```

```python
df['6 Clusters'] = y_pred
df
```

```python
df['6 Clusters'].value_counts()
```

```python
def make_cluster_plot(data, n_clusters, x, y, cluster_column):
    for cluster in range(n_clusters):
        tdf = data[data[cluster_column] == cluster]
        plt.scatter(tdf[x], tdf[y], label=cluster)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
```

```python
make_cluster_plot(df, 6, 'Asteroid Velocity', 'Possible Impacts', '6 Clusters')
```

```python
make_cluster_plot(df, 6, 'Asteroid Magnitude', 'Possible Impacts', '6 Clusters')
```

```python
make_cluster_plot(df, 6, 'Maximum Palermo Scale', 'Possible Impacts', '6 Clusters')
```

```python
model.cluster_centers_
```

Now our main aim was to see when category has maximum impact so we can take the cluster center values for each and see which has maximum impact

```python
centers = model.cluster_centers_
```

```python
possible_impact_centers = []
for i in range(len(centers)):
    possible_impact_centers.append((i, centers[i][2]))
possible_impact_centers.sort(key = lambda x: x[1]) 
possible_impact_centers
```

So the cluster with maximum impact is 2.
The order of impact is 2, 5, 0, 4, 3, 1

```python
cluster_impact_mapping = {2:5, 5:4, 0:3, 4:2, 3:1, 1:0}
```

```python
df['6 Clusters'].replace(cluster_impact_mapping, inplace=True)
```

```python
df
```

Hence now these are the clusters based on their possible impact using the given features which can be useful to classify the asteroids in future based on the given features. 0 Begin the least harmful and 5 being the maximum harmful.
